```python
try:
    # Connect to SingleStore database
    connection = mysql.connector.connect(
        host=host,
        user=user,
        password=password,
        database=database,
        port=port
    )
    
    if connection.is_connected():
        print("Connection to SingleStore was successful.")
        
        cursor = connection.cursor()

        # Execute SQL query
        query = "SELECT * FROM campaigns"
        cursor.execute(query)

        # Fetch all results
        results = cursor.fetchall()

        # Get column names
        columns = [i[0] for i in cursor.description]

        # Open a CSV file for writing
        with open('campaigns_data.csv', mode='w', newline='', encoding='utf-8') as file:
            writer = csv.writer(file)

            # Write column names as the first row
            writer.writerow(columns)

            # Write the results to the CSV
            writer.writerows(results)

        print("Data has been successfully saved to 'campaigns_data.csv'.")

        # Execute SQL query
        query = "SELECT * FROM events"
        cursor.execute(query)

        # Fetch all results
        results = cursor.fetchall()

        # Get column names
        columns = [i[0] for i in cursor.description]

        # Open a CSV file for writing
        with open('events_data.csv', mode='w', newline='', encoding='utf-8') as file:
            writer = csv.writer(file)

            # Write column names as the first row
            writer.writerow(columns)

            # Write the results to the CSV
            writer.writerows(results)

        print("Data has been successfully saved to 'events_data.csv'.")

except Error as e:
    print(f"Error while connecting to SingleStore: {e}")

finally:
    if connection.is_connected():
        cursor.close()
        connection.close()
        print("Connection closed.")
```
