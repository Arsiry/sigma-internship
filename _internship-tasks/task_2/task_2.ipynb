The goal of digital advertising is to inorganically advertise where consumers are and to customize ads to the target audience's preferences.

### Machine Learning

#### Decision Trees

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
```

```python
# Split Data
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 1)
```

```python
# Specify the model. 
# For model reproducibility, set a numeric value for random_state when specifying the model
model_tree = DecisionTreeRegressor(random_state=1)

# Fit the model with the training data.
model_tree.fit(train_X, train_y)
```

```python
# Make Predictions with Validation data
# Predict with all validation observations
y_pred = model_tree.predict(val_X)

print("Accuracy is %2.3f" % mean_absolute_error(val_y, y_pred))
```

```python
def get_score(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=1)
    model.fit(train_X, train_y)
    y_pred = model.predict(val_X)
    return(mean_absolute_error(val_y, y_pred))
```

```python
# Compare Different Tree Sizes
candidate_max_leaf_nodes = [5, 25, 50, 100, 250, 500, 1000, 1500, 2000, 3000, 5000, 7500, 10000, 15000, 20000, 30000, 50000, 100000]
# Write loop to find the ideal tree size from candidate_max_leaf_nodes
for max_leaf_nodes in candidate_max_leaf_nodes:
    my_score = get_score(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d  \t\t Accuracy is %2.3f" %(max_leaf_nodes, my_score))
```

#### Random Forest

```python
# Define the model. 
rf_model = RandomForestRegressor(random_state=1)

# Fit your model
rf_model.fit(train_X, train_y)
```

```python
# Make Predictions with Validation data
# Predict with all validation observations
y_pred = rf_model.predict(val_X)

print("Accuracy is %2.3f" % mean_absolute_error(val_y, y_pred))
```
